```python
#### This code is for ATOC7500 project 4 part I
#import modules
import numpy as np   #for calculations
import matplotlib.pyplot as plt  #for plotting
%matplotlib inline  
import pandas as pd  #for reading in data
import datetime    #for manipulation by dtae
import collections  #for additional dictionary functionality
from matplotlib.gridspec import GridSpec #for subplots
import scipy.signal as signal #for signal processing
import scipy.stats as stats
import xarray
import netCDF4 as nc
```

```python
def create_normalized_redfit(data_length,Te):
    freq = np.arange(0,(data_length/2)+1,1)/float(data_length) # to Nyquist
    red_fit = (2 * Te)/(1 + ((2*np.pi*freq)**2)*(Te**2)) # After Hartmann 6.64, 6.91
    return red_fit/np.sum(red_fit)
def create_f_bounds(alpha,dof,red_fit_n):
    f_ratio = stats.f.ppf(alpha,dof,200) # Note: 200 = large degree of freedom for red noise
    return f_ratio*red_fit_n
```

```python
## read in data
range1=700*12   #for first 700 years
range2=100*12   #for first 100 years

nc_fid1=nc.Dataset('coupled.nc')
nc_fid2=nc.Dataset('SOM.nc')

data_coupled_700_temp=nc_fid1['nino34'][0:range1]
data_coupled_700=data_coupled_700_temp-np.mean(data_coupled_700_temp)
data_coupled_100_temp=nc_fid1['nino34'][0:range2]
data_coupled_100=data_coupled_100_temp-np.mean(data_coupled_100_temp)

data_som_700_temp=nc_fid2['nino34'][0:range1]
data_som_700=data_som_700_temp-np.mean(data_som_700_temp)
data_som_100_temp=nc_fid2['nino34'][0:range2]
data_som_100=data_som_100_temp-np.mean(data_som_100_temp)

print('length of first 700 years is:', len(data_coupled_700), 'months')
print('length of first 100 years is:', len(data_coupled_100), 'months')
```

```python
### Calculate lag-1 autocorrelation and e-folding time of data
rho_c_700,lag1_p = stats.pearsonr(data_coupled_700[0:range1-1],data_coupled_700[1:range1]) # rho = lag1_r
Te_c_700 = -1./np.log(rho_c_700) # After Hartmann 6.4 with delta t = 1
print('coupled 700, autocorrelation=',round(rho_c_700,2),'and Te =',round(Te_c_700,2),'months')
print('  ')

rho_s_700,lag1_p = stats.pearsonr(data_som_700[0:range1-1],data_som_700[1:range1]) # rho = lag1_r
Te_s_700 = -1./np.log(rho_s_700) # After Hartmann 6.4 with delta t = 1
print('slab ocean 700, autocorrelation=',round(rho_s_700,2),'and Te =',round(Te_s_700,2),'months')
print('  ')

rho_c_100,lag1_p = stats.pearsonr(data_coupled_100[0:range2-1],data_coupled_100[1:range2]) # rho = lag1_r
Te_c_100 = -1./np.log(rho_c_100) # After Hartmann 6.4 with delta t = 1
print('coupled 100, autocorrelation=',round(rho_c_100,2),'and Te =',round(Te_c_100,2),'months')
print('  ')

rho_s_100,lag1_p = stats.pearsonr(data_som_100[0:range2-1],data_som_100[1:range2]) # rho = lag1_r
Te_s_100 = -1./np.log(rho_s_100) # After Hartmann 6.4 with delta t = 1
print('slab ocean 100, autocorrelation=',round(rho_s_100,2),'and Te =',round(Te_s_100,2),'months')
```

```python
## Use the Hanning Window AND change the length of the data window AND use WOSA (Welch’s Overlapping Segment Analysis)
window_length=50*12   # 50 years into months
T2 = window_length/2
freq_w = np.arange(0.,T2+1.)/window_length
P_welch1 = signal.welch(data_coupled_700,window='hanning',nperseg=window_length);
P_welch_n1 = P_welch1[1]/np.sum(P_welch1[1])  #normalize
P_welch2 = signal.welch(data_som_700,window='hanning',nperseg=window_length);
P_welch_n2 = P_welch2[1]/np.sum(P_welch2[1])  #normalize

## Construct the expected rednoise spectrum - functions
## Need to update the 99% confidence interval
alpha = 0.99
red_fit_n_welch1 = create_normalized_redfit(window_length,Te_c_700)  ## generate red_fit_n using window_length
red_fit_n_welch2 = create_normalized_redfit(window_length,Te_s_700) 
dof_welch=len(data_coupled_700)/(window_length/2)  ### Barnes Eq. 26
f_bounds_welch1 = create_f_bounds(alpha,dof_welch,red_fit_n_welch1)  ## f-test for variance; 
f_bounds_welch2 = create_f_bounds(alpha,dof_welch,red_fit_n_welch2)

## plot
fig = plt.figure(figsize=(14,4))
plt.xlabel('Frequency')
plt.ylabel('Normalized Power')
plt.ylim(0.0,0.2)
plt.xlim(0.0,0.05)
plt.grid()
plt.plot(freq_w,P_welch_n1, color='blue', lw=1.3, label='Hanning for Coupled Data,wl='+str(window_length));
plt.plot(freq_w,P_welch_n2, color='orange', lw=1.3, label='Hanning for SOM Data,wl='+str(window_length));
plt.plot(freq_w,red_fit_n_welch1, color='blue', lw=1, alpha=0.9, label='Red Noise Fit for Coupled Data');
plt.plot(freq_w,f_bounds_welch1, color='blue', lw=1, ls='--', alpha=0.9, label=str(int(alpha*100))+'% CI');
plt.plot(freq_w,red_fit_n_welch2, color='orange', lw=1, alpha=0.9, label='Red Noise Fit for SOM Data');
plt.plot(freq_w,f_bounds_welch2, color='orange', lw=1, ls='--', alpha=0.9, label=str(int(alpha*100))+'% CI');
plt.title("Normalized Power Spectrum of Pre-industrial Control Run of CESM Large Ensemble \n WOSA, Hanning window using window length="+str(window_length)+",DOF="+str(round(dof_welch,0)),fontsize=14);
plt.legend(loc='upper right',fontsize=14)
fig.tight_layout()
```

```python
# Find the significant spectral peaks
print('peaks of coupled 700 years data:')
print('   ')
for i in range(len(P_welch_n1)):
    if (P_welch_n1[i] - f_bounds_welch1[i]) > 5e-03:
        print('##### FOUND IT - spectral peak exceeds red noise ####')
        print('exceeds by...',P_welch_n1[i]-f_bounds_welch1[i])
        print('at frequency....',freq_w[i])
        print('which in months is approximately...',round((1/freq_w[i]))) ## use the regridded regular spacing   
        print('    ')
        print('    ')
```
